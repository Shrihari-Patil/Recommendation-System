```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('book.csv',encoding = 'latin-1')
df.head()
```

```python
book = df.drop(columns = 'Unnamed: 0')
book = book.rename(columns = {'User.ID':'user_id','Book.Title':'book_title','Book.Rating':'book_rating'})
book.head()
```

```python
book.info()
```

```python
avg_rating = book.book_rating.mean()
avg_rating
```

```python
#No of unique user's in the dataset
len(book.user_id.unique())
```

```python
#No of unique book_title in the dataset
len(book.book_title.unique())
```

Pivoting Dataset

```python
book_df = book.pivot_table(index = 'user_id', columns = 'book_title', values = 'book_rating').reset_index(drop = True)
book_df.index = book.user_id.unique()
book_df.head()
```

```python
#fill null values to zero
book_df.fillna(0, inplace = True)
book_df.head()
```

# Cosine Similarity

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( book_df.values,metric = 'cosine')
```

```python
user_sim_df = pd.DataFrame(user_sim)
user_sim_df.index = book.user_id.unique()
user_sim_df.columns = book.user_id.unique()
user_sim_df.iloc[:10,:10]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[:10,:10]
```

```python
# Similar Users
user_sim_df.idxmax(axis = 1)[0:10]
```

```python
book_user_rating = book_df["You Don't Need Experience if You've Got Attitude"]
book_user_rating
```

```python
# Correlation with other Books
similar_book = book_df.corrwith(book_user_rating)
corr_book = pd.DataFrame(similar_book, columns = ['Correlation'])
corr_book.dropna(inplace = True)
corr_book.head()
```
